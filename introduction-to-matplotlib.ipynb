```python
import numpy as np
# Create Some Kind of Data
x=np.linspace(0,10,100)
```

```python
# Making plots with dictionary
butter = {"Almond Price": 10, "Peanut Price": 8, "Chehew Price": 12}
fig, ax = plt.subplots()
ax.bar(butter.keys(), butter.values())
ax.set(title="Butter Prices", xlabel="Butter Name", ylabel="Price in Dollars");
```

### Which one to use for plotting(pyplot vs matplotlib OO method)
* For quick plot pyplot is better
* For more advanced options matplotlib is a good option

```python
heart_rate.head()
```

```python
over_50 = heart_rate[heart_rate["age"] > 50]
over_50.head()
```

```python
# Simple yplot method
over_50.plot(kind='scatter', x="age",y="chol",c="target");
```

```python
# OO method mixed with pyplot method
fig,  ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter', x="age",y="chol",c="target",ax=ax);
```

```python
# ## OO method 
# fig, ax = plt.subplots(figsize=(10,6))
# scatter = ax.scatter(x=over_50["age"],
#                       y=over_50["chol"],
#                       c=over_50["target"])
# ax.set(title="Heart and Cholestrol level",
#       xlabel="Age",
#       ylabel="Cholestrol LeveL")
# # Add some legend

# ax.legend(*scatter.legend_elements(), title="Target");
# # Add a horizontal Lines
# ax.axhline(over_50["chol"].mean(),linestyle='--')
```

```python
## OO method 
fig, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c = over_50["target"])
# Assigning Values
ax.set(title="Heart and Cholestrol",xlabel="Age",ylabel="Cholestrol")
## Setting target / Legend
ax.legend(*scatter.legend_elements(), title="Target");

# Adding an Horizontal Line
ax.axhline(over_50["chol"].mean(), linestyle="--")
```

```python
over_50.head()
```

```python
# Subplot of age, chol and thalach
```

```python
fig, (ax0, ax1)= plt.subplots(nrows=2,ncols=1,figsize=(10,10))
# Add some data to ax0
scatter = ax0.scatter(x=over_50["age"],y=over_50["chol"], c=over_50["target"])
# Set some titles
ax0.set(title="Heart Disease",xlabel="Age",ylabel="Cholestrol",)
# Add Some lagend in our Data Frame
ax0.legend(*scatter.legend_elements(),title="Target");
# Add some mean line
ax0.axhline(over_50["chol"].mean(),linestyle="--")


scatter1 = ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"])
ax1.set(title="Heart Diseases", xlabel="Age",ylabel="Thalach")
ax1.legend(*scatter1.legend_elements(),title="Target")
ax1.axhline(over_50["thalach"].mean(),linestyle="--");

# Add complete title on this figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```
